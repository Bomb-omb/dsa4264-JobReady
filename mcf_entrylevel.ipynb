```python
import re

def remove_emojis(text):
    emoji_pattern = re.compile(
        "["                       # start character class
        "\U0001F600-\U0001F64F"   # emoticons
        "\U0001F300-\U0001F5FF"   # symbols & pictographs
        "\U0001F680-\U0001F6FF"   # transport & map symbols
        "\U0001F1E0-\U0001F1FF"   # flags
        "\U00002700-\U000027BF"   # dingbats
        "\U000024C2-\U0001F251"
        "]+",
        flags=re.UNICODE
    )
    return emoji_pattern.sub(r'', str(text))

# clean the description column
df['description'] = df['description'].apply(remove_emojis)
```

# Check that each extracted skill match the parent skill title in skill taxonomy
It’s called a `token` here because each piece of extracted_skills is being treated as one split unit after you break the string on |.

```python
import pandas as pd


def _solve_skill_token_sequence(tokens, taxonomy_titles, max_join=6):
    token_count = len(tokens)
    edges = {start: [(start + 1, 0, None)] for start in range(token_count)}

    for start in range(token_count):
        parts = []
        for end in range(start, min(token_count, start + max_join)):
            parts.append(tokens[end])
            candidate = ", ".join(parts)
            if candidate in taxonomy_titles:
                span = end - start + 1
                edges[start].append((end + 1, span, candidate))

    best_from = [0] * (token_count + 1)
    for start in range(token_count - 1, -1, -1):
        best_from[start] = max(weight + best_from[end] for end, weight, _ in edges[start])

    optimal_edges = {start: [] for start in range(token_count)}
    for start in range(token_count):
        for end, weight, candidate in edges[start]:
            if weight + best_from[end] == best_from[start]:
                optimal_edges[start].append((end, weight, candidate))

    ways_prefix = [0] * (token_count + 1)
    ways_prefix[0] = 1
    for start in range(token_count):
        for end, _, _ in optimal_edges[start]:
            ways_prefix[end] += ways_prefix[start]

    ways_suffix = [0] * (token_count + 1)
    ways_suffix[token_count] = 1
    for start in range(token_count - 1, -1, -1):
        ways_suffix[start] = sum(ways_suffix[end] for end, _, _ in optimal_edges[start])

    total_optimal_paths = ways_suffix[0]
    guaranteed_segments = []
    covered_positions = set()

    for start in range(token_count):
        for end, weight, candidate in optimal_edges[start]:
            if weight == 0:
                continue
            if ways_prefix[start] * ways_suffix[end] == total_optimal_paths:
                guaranteed_segments.append((start, end, candidate))
                covered_positions.update(range(start, end))

    guaranteed_segments.sort(key=lambda item: (item[0], item[1]))
    unresolved_positions = [pos for pos in range(token_count) if pos not in covered_positions]
    return guaranteed_segments, unresolved_positions


def build_skill_audit_report(
    mcf_path,
    taxo_path,
    extracted_col="extracted_skills",
    taxo_col="parent_skill_title",
    delimiter="|",
    max_join=6,
):
    jobs_df = pd.read_csv(mcf_path, encoding="utf-8-sig")
    taxo_df = pd.read_csv(taxo_path, encoding="utf-8-sig")

    taxonomy_titles = {
        str(value).strip()
        for value in taxo_df[taxo_col].dropna()
        if str(value).strip()
    }

    raw_token_rows = []
    normalized_rows = []
    manual_review_rows = []
    unique_raw_tokens = set()
    unique_unresolved_tokens = set()

    non_empty_mask = jobs_df[extracted_col].fillna("").astype(str).str.strip().ne("")

    for idx, row in jobs_df.loc[non_empty_mask].iterrows():
        raw_text = str(row[extracted_col]).strip()
        tokens = [part.strip() for part in raw_text.split(delimiter) if part.strip()]
        row_number = idx + 2
        row_id = row.get("uuid", "")
        row_title = row.get("title", "")

        for token in tokens:
            unique_raw_tokens.add(token)
            raw_token_rows.append(
                {
                    "row_number": row_number,
                    "uuid": row_id,
                    "title": row_title,
                    "raw_extracted_skills": raw_text,
                    "raw_skill": token,
                    "raw_exact_match": token in taxonomy_titles,
                }
            )

        guaranteed_segments, unresolved_positions = _solve_skill_token_sequence(
            tokens,
            taxonomy_titles,
            max_join=max_join,
        )

        for start, end, candidate in guaranteed_segments:
            normalized_rows.append(
                {
                    "row_number": row_number,
                    "uuid": row_id,
                    "title": row_title,
                    "raw_extracted_skills": raw_text,
                    "normalized_skill": candidate,
                    "source_tokens": f" {delimiter} ".join(tokens[start:end]),
                    "resolution_type": "exact" if end - start == 1 else "recombined",
                }
            )

        if unresolved_positions:
            unresolved_tokens = [tokens[pos] for pos in unresolved_positions]
            unique_unresolved_tokens.update(unresolved_tokens)
            manual_review_rows.append(
                {
                    "row_number": row_number,
                    "uuid": row_id,
                    "title": row_title,
                    "raw_extracted_skills": raw_text,
                    "unresolved_tokens": " | ".join(unresolved_tokens),
                    "resolution_type": "unresolved",
                }
            )

    raw_token_df = pd.DataFrame(raw_token_rows)
    normalized_token_df = pd.DataFrame(normalized_rows)
    manual_review_df = pd.DataFrame(manual_review_rows)

    summary = {
        "mcf_rows": len(jobs_df),
        "rows_with_extracted_skills": int(non_empty_mask.sum()),
        "taxonomy_parent_titles": len(taxonomy_titles),
        "unique_raw_tokens": len(unique_raw_tokens),
        "raw_exact_matches": sum(token in taxonomy_titles for token in unique_raw_tokens),
        "raw_exact_mismatches": sum(token not in taxonomy_titles for token in unique_raw_tokens),
        "unique_normalized_titles": int(normalized_token_df["normalized_skill"].nunique()) if not normalized_token_df.empty else 0,
        "recombined_token_count": int((normalized_token_df["resolution_type"] == "recombined").sum()) if not normalized_token_df.empty else 0,
        "manual_review_row_count": len(manual_review_df),
        "manual_review_token_count": len(unique_unresolved_tokens),
    }

    return {
        "summary": summary,
        "raw_token_df": raw_token_df,
        "normalized_token_df": normalized_token_df,
        "manual_review_df": manual_review_df,
    }
```

```python
report = build_skill_audit_report(
    "data/mcf_entrylevel.csv",
    "data/skills_taxo.csv",
)

summary_df = pd.DataFrame(
    report["summary"].items(),
    columns=["metric", "value"],
)

raw_mismatch_df = (
    report["raw_token_df"]
    .loc[lambda df: ~df["raw_exact_match"]]
    .groupby("raw_skill", as_index=False)
    .size()
    .rename(columns={"size": "frequency"})
    .sort_values(["frequency", "raw_skill"], ascending=[False, True])
    .reset_index(drop=True)
)

recombined_df = (
    report["normalized_token_df"]
    .loc[lambda df: df["resolution_type"].eq("recombined")]
    .groupby(["source_tokens", "normalized_skill"], as_index=False)
    .size()
    .rename(columns={"source_tokens": "raw_skill", "size": "frequency"})
    .sort_values(["frequency", "normalized_skill", "raw_skill"], ascending=[False, True, True])
    .reset_index(drop=True)
)


manual_review_df = (
    report["manual_review_df"]
    .sort_values(["row_number", "title"], ascending=[True, True])
    .reset_index(drop=True)
)
```

## Some skill were split by `|` when it shouldn't, so we will remove it

This happens because the skills future website used `,` in the name of a skill, so the skill name got seperated when downloading the csv from the website

```python
import pandas as pd


def _solve_skill_token_sequence(tokens, taxonomy_titles, max_join=6):
    token_count = len(tokens)
    edges = {start: [(start + 1, 0, None)] for start in range(token_count)}

    for start in range(token_count):
        parts = []
        for end in range(start, min(token_count, start + max_join)):
            parts.append(tokens[end])
            candidate = ", ".join(parts)
            if candidate in taxonomy_titles:
                span = end - start + 1
                edges[start].append((end + 1, span, candidate))

    best_from = [0] * (token_count + 1)
    for start in range(token_count - 1, -1, -1):
        best_from[start] = max(weight + best_from[end] for end, weight, _ in edges[start])

    optimal_edges = {start: [] for start in range(token_count)}
    for start in range(token_count):
        for end, weight, candidate in edges[start]:
            if weight + best_from[end] == best_from[start]:
                optimal_edges[start].append((end, weight, candidate))

    ways_prefix = [0] * (token_count + 1)
    ways_prefix[0] = 1
    for start in range(token_count):
        for end, _, _ in optimal_edges[start]:
            ways_prefix[end] += ways_prefix[start]

    ways_suffix = [0] * (token_count + 1)
    ways_suffix[token_count] = 1
    for start in range(token_count - 1, -1, -1):
        ways_suffix[start] = sum(ways_suffix[end] for end, _, _ in optimal_edges[start])

    total_optimal_paths = ways_suffix[0]
    guaranteed_segments = []
    covered_positions = set()

    for start in range(token_count):
        for end, weight, candidate in optimal_edges[start]:
            if weight == 0:
                continue
            if ways_prefix[start] * ways_suffix[end] == total_optimal_paths:
                guaranteed_segments.append((start, end, candidate))
                covered_positions.update(range(start, end))

    guaranteed_segments.sort(key=lambda item: (item[0], item[1]))
    unresolved_positions = [pos for pos in range(token_count) if pos not in covered_positions]
    return guaranteed_segments, unresolved_positions


def normalize_extracted_skills(raw_text, taxonomy_titles, delimiter="|", max_join=6):
    if pd.isna(raw_text) or not str(raw_text).strip():
        return "", [], False

    raw_text = str(raw_text).strip()
    tokens = [part.strip() for part in raw_text.split(delimiter) if part.strip()]

    guaranteed_segments, unresolved_positions = _solve_skill_token_sequence(
        tokens,
        taxonomy_titles,
        max_join=max_join,
    )

    segment_by_start = {start: (end, candidate) for start, end, candidate in guaranteed_segments}

    normalized_parts = []
    pos = 0
    while pos < len(tokens):
        if pos in segment_by_start:
            end, candidate = segment_by_start[pos]
            normalized_parts.append(candidate)
            pos = end
        else:
            normalized_parts.append(tokens[pos])
            pos += 1

    normalized_text = " | ".join(normalized_parts)
    unresolved_tokens = [tokens[pos] for pos in unresolved_positions]
    changed = normalized_text != raw_text

    return normalized_text, unresolved_tokens, changed


mcf_path = "data/mcf_entrylevel.csv"
taxo_path = "data/skills_taxo.csv"

jobs_df = pd.read_csv(mcf_path, encoding="utf-8-sig")
taxo_df = pd.read_csv(taxo_path, encoding="utf-8-sig")

taxonomy_titles = {
    str(value).strip()
    for value in taxo_df["parent_skill_title"].dropna()
    if str(value).strip()
}

normalized_values = []
manual_review_rows = []
changed_rows = 0

for idx, row in jobs_df.iterrows():
    raw_value = row.get("extracted_skills", "")
    normalized_text, unresolved_tokens, changed = normalize_extracted_skills(
        raw_value,
        taxonomy_titles,
        delimiter="|",
        max_join=6,
    )

    if pd.isna(raw_value) or not str(raw_value).strip():
        normalized_values.append("")
    else:
        normalized_values.append(normalized_text)

    if changed:
        changed_rows += 1

    if unresolved_tokens:
        manual_review_rows.append(
            {
                "row_number": idx + 2,
                "uuid": row.get("uuid", ""),
                "title": row.get("title", ""),
                "unresolved_tokens": " | ".join(unresolved_tokens),
                "normalized_extracted_skills": normalized_text,
            }
        )

jobs_df["extracted_skills"] = normalized_values
jobs_df.to_csv(mcf_path, index=False, encoding="utf-8-sig")

manual_review_df = (
    pd.DataFrame(manual_review_rows)
    .sort_values(["row_number", "title"], ascending=[True, True])
    .reset_index(drop=True)
)

print(f"Updated file in place: {mcf_path}")
print(f"Rows changed: {changed_rows}")

```
