### Filter entry level and jobs with bachelor's degree

```python
keywords = r"bachelor|degree|graduate|university|undergraduate|tertiary"
levels = r'"Fresh/entry level"|"Junior Executive"'

mcf_filtered1 = mcf[
    (
        (mcf["minimumYearsExperience"] == 0) |
        (mcf["positionLevels"].str.contains(levels, case=False, na=False))
    )
    &
    (mcf["description"].str.contains(keywords, case=False, na=False))
     & 
    (mcf["minimumYearsExperience"]<4)
   
]

mcf_filtered1.head()
```

### Save dataset

```python
mcf_filtered1.to_csv("data/mcf_entrylevel.csv", index=False)
```

### 5 X 50 random

```python
import pandas as pd

# sample 250 rows
sampled = mcf_filtered1.sample(n=250, random_state=42).copy()

# assign numbers 1–5 (50 rows each)
assignees = (
    [1]*50 +
    [2]*50 +
    [3]*50 +
    [4]*50 +
    [5]*50
)

sampled["assignee"] = assignees

# optional columns for your team
sampled["extracted_skills"] = ""
sampled["done"] = ""

sampled.to_csv("data/mcf_sample_250_assigned.csv", index=False)

sampled.head()
```

```python
for i in range(1, 6):
    sampled[sampled["assignee"] == i].to_csv(
        f"data/jd{i}.csv", index=False
    )
```
