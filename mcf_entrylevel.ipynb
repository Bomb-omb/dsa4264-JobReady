### Filter to Entry level & Junior Executive only

```python
levels = r"\bFresh/entry level\b|\bJunior Executive\b"

mcf_filtered = mcf[
    mcf["positionLevels"].str.contains(levels, case=False, na=False)
]

mcf_filtered.head()
```

### Number of rows and column type

```python
len(mcf_filtered)
```

```python
mcf_filtered["positionLevels"].unique()
```

```python
keywords = r"bachelor|degree|graduate|university|undergraduate|tertiary"


mcf_filtered1 = mcf[
    mcf["description"].str.contains(keywords, na=False)
]

mcf_filtered1.head()
```

```python
len(mcf_filtered1)
```

### Save dataset

```python
mcf_filtered.to_csv("data/mcf_entrylevel.csv", index=False)
```
